This project explores the classic Iris dataset using Python. It includes data cleaning, descriptive statistics, and visualizations such as histograms and pair plots to uncover patterns across species. The goal is to demonstrate foundational EDA techniques and build intuition for multivariate data.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("iris.csv")
print(df.head())
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df.hist(figsize=(8,6))
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=df)
plt.title("Boxplot of Iris Features")
plt.show()
```

```python
sns.pairplot(df, hue="species")
plt.suptitle("Pairplot of Iris Features by Species", y=1.02)
plt.show()
```

```python
df.groupby("species")["petal_length"].mean()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

# EDA on Iris Dataset

## 1. Introduction
Exploratory Data Analysis (EDA) on the classic Iris dataset to understand feature distributions, relationships, and species differences.

## 2. Data Overview
- 150 rows, 5 features + species label
- No missing values
- Numeric features: sepal length/width, petal length/width

## 3. Visualizations
- Histograms show feature distributions
- Boxplots reveal outliers and spread
- Pairplot highlights species separation

## 4. Key Insights
- Virginica has the longest petals on average
- Petal length and width are highly correlated
- Setosa is clearly separable from other species

## 5. Conclusion
This EDA provides a strong foundation for classification tasks and species prediction. Visual patterns suggest petal dimensions are key discriminators.
