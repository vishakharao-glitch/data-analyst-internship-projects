This project analyzes historical stock prices with a focus on Apple Inc. It includes time series visualization using Python and a Power BI dashboard to explore trends, moving averages, and trading volumes. The goal is to demonstrate time series analysis and storytelling through data.


```python
import pandas as pd

# Load the dataset
df = pd.read_csv("stock_prices.csv", low_memory=False)

# Preview the first few rows and column names
print(df.head())
print(df.columns)
```

```python
# Filter for Apple stock
apple_df = df[df["symbol"] == "AAPL"]

# Sort by date
apple_df = apple_df.sort_values("date")

# Preview
print(apple_df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(apple_df["date"], apple_df["close"], label="Close Price")
plt.title("Apple Stock Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate 7-day and 30-day moving averages
apple_df["MA_7"] = apple_df["close"].rolling(window=7).mean()
apple_df["MA_30"] = apple_df["close"].rolling(window=30).mean()

# Plot closing price with moving averages
import matplotlib.pyplot as plt

plt.figure(figsize=(14,7))
plt.plot(apple_df["date"], apple_df["close"], label="Close Price", color="blue")
plt.plot(apple_df["date"], apple_df["MA_7"], label="7-Day MA", color="orange")
plt.plot(apple_df["date"], apple_df["MA_30"], label="30-Day MA", color="green")
plt.title("Apple Stock Price with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### ðŸ“Š Insights from Apple Stock Analysis

- The 7-day moving average captures short-term fluctuations, useful for active traders.
- The 30-day moving average smooths out noise and reveals long-term trends.
- Appleâ€™s stock shows a steady upward trend post-Q1 earnings, with occasional dips in volume.
- These patterns can inform buy/sell decisions and risk management strategies.
